```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
def logistic4(x, A, B, C, D, G):
    """4PL logistic equation."""
    return D + (A-D)/((1 + (x/C)**B)**G)

A = 3520
B = 1.2
C = 3020
D = 2570000
G = 1

xCont = np.linspace(0.01, 1000000, 1000000)
fittedData=[logistic4(xVal, A, B, C, D, G) for xVal in xCont]
fig1 = plt.figure(figsize=(10,6))
ax=fig1.add_subplot(1,1,1)
ax.plot(xCont, fittedData)
ax.set_ylabel("y axis")
ax.set_xlabel("""x axis

5-P Fit: y = D+(A-D)/((1+(x/C)^B)^G)
A:{0:0.3g}, B:{1:0.3g}, C:{2:0.3g}, D:{3:0.3g}, G:{4:0.3g}
""".format(A, B, C, D, G))
ax.grid(True, which='both')
ax.set_title("Test plot")
plt.xscale('log')
plt.axis([0.1, 1000000, .5, 2600000])
```
